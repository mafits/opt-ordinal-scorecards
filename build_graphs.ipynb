```python
def acc_vs_size_plot(dataset_name, accuracies, model_sizes):
    model_names = ['Logistic Regression', 'SVM', 'Adaptive Lasso', 'skglm']
    colors = ['red', 'green', 'blue', 'orange']

    '''
    # RGB values corresponding to the colors
    rgb_colors = [(1, 0, 0),      # red
                (0, 1, 0),      # green
                (0, 0, 1),      # blue
                (1, 0.65, 0)]   # orange

    '''

    plt.figure(figsize=(10, 6))
    plt.scatter(model_sizes, accuracies, color=colors, s=150)

    x_limits = plt.xlim()
    y_limits = plt.ylim()

    # add dashed lines from each point to the axes
    for x, y, c in zip(model_sizes, accuracies, colors):
        plt.axhline(y, xmin=0, xmax=(x - x_limits[0]) / (x_limits[1] - x_limits[0]),
                    color=c, linestyle='--', linewidth=1, alpha=1)
        plt.axvline(x, ymin=0, ymax=(y - y_limits[0]) / (y_limits[1] - y_limits[0]),
                    color=c, linestyle='--', linewidth=1, alpha=1)

    plt.title(dataset_name)
    plt.xlabel('Model Size')
    plt.ylabel('Test Accuracy')
    plt.grid()
    plt.show()
```
